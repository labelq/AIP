# Деплоймент

Поздравляю, если вы дошли до этого шага, то половина зачетного минимума уже у вас есть (работающий бот с TinyDB). Сейчас научимся деплоить проект на сервер. Для этого:


1.   Регистрируем бесплатный аккаунт https://www.pythonanywhere.com/
2.   Заходим в Files и жмем Open Bash console here


image.png

3. Клонируете вашего бота из гит хаб через git clone

4. В файле requirements.txt должны быть все ваши библиотеки.
В открывшейся консоли устанавливаем зависимости по файлу.

`pip install -r ci-cd/requirements.txt`

image.png

5. Запускаем бота

`python3 test_AIP/src/app.py`

Могут возникнуть ошибки импортов, или индексов. Убедитесь, что внесли все зависимости в requirements и ваша БД инициализируется в нужной папке.

После запуска бот обычно работает от 3 до 24 часов в зависимости от нагрузки на сервера сайта. 

# Зачетный минимум

1.   Код бота оформленный по инструкции с лекций/гитхаба (обязательно с БД и кнопками, как в примерах)
2.   Бот должен быть размещен на сервере и запущен (работать без ошибок в телеграме)
3. Вы должны понимать, что написали (как минимум ответить на вопрос что будет, если удалить ту или иную строчку)
